### Zero-Shot Prompting with Ollama (llama2)

**Concept**: In zero-shot prompting, the model is expected to perform a task without any examples, relying solely on its pre-trained knowledge to understand the instruction and generate a relevant response.

**How it works with Ollama**: We will send a direct prompt to the `llama2` model through the Ollama client, asking it to perform a task. The model uses its extensive training data to understand the request and generate an appropriate output.

```python
zero_shot_prompt_ollama = "Translate the following English sentence to French: 'Hello, how are you?'"

# Call Ollama model
response = ollama.chat(model='llama2', messages=[{
    'role': 'user',
    'content': zero_shot_prompt_ollama,
}])
zero_shot_output_ollama = response['message']['content']

print(f"--- Prompt ---\n{zero_shot_prompt_ollama}\n")
print(f"--- Model Output (llama2) ---\n{zero_shot_output_ollama}")
```

**Explanation of output (llama2)**: The `llama2` model, through Ollama, translates the sentence directly based on the zero-shot instruction. The quality of the translation will depend on the model's capabilities without specific examples.

### One-Shot Prompting with Ollama (llama2)

**Concept**: The model is given *one example* of the task along with the task description. This single example guides the model towards the desired output format, style, or specific interpretation of the instruction.

**How it works with Ollama**: We'll construct a prompt that includes one input-output example for the `llama2` model to learn from, helping it understand the desired format for the subsequent task.

```python
one_shot_prompt_ollama = (
    "Classify the sentiment of the following text as Positive, Negative, or Neutral.\n\n"
    "Text: 'I love this product!'\n"
    "Sentiment: Positive\n\n"
    "Text: 'This movie was okay, not great.'\n"
    "Sentiment:"
)

# Call Ollama model
response = ollama.chat(model='llama2', messages=[{
    'role': 'user',
    'content': one_shot_prompt_ollama,
}])
one_shot_output_ollama = response['message']['content']

print(f"--- Prompt ---\n{one_shot_prompt_ollama}\n")
print(f"--- Model Output (llama2) ---\n{one_shot_output_ollama}")
```

**Explanation of output (llama2)**: By providing one example, the `llama2` model is guided to classify the sentiment in the specified format, demonstrating how a single example can influence the model's understanding of the task.

### Few-Shot Prompting with Ollama (llama2)

**Concept**: Similar to one-shot, but the model is provided with *several examples* (a "few shots") of the task. This can significantly improve performance for more complex tasks, when the desired output format is nuanced, or when the model needs to learn a specific mapping from inputs to outputs.

**How it works with Ollama**: We'll provide multiple input-output pairs in the prompt to `llama2`, allowing it to better generalize the pattern and perform the task more accurately.

```python
few_shot_prompt_ollama = (
    "Extract the city and state from the following addresses:\n\n"
    "Address: 123 Main St, Anytown, CA 90210\n"
    "City: Anytown, State: CA\n\n"
    "Address: 456 Oak Ave, Big City, NY 10001\n"
    "City: Big City, State: NY\n\n"
    "Address: 789 Pine Ln, Smallville, TX 75001\n"
    "City: Smallville, State: TX\n\n"
    "Address: 101 Elm Rd, Metropolis, IL 60601\n"
    "City:"
)

# Call Ollama model
response = ollama.chat(model='llama2', messages=[{
    'role': 'user',
    'content': few_shot_prompt_ollama,
}])
few_shot_output_ollama = response['message']['content']

print(f"--- Prompt ---\n{few_shot_prompt_ollama}\n")
print(f"--- Model Output (llama2) ---\n{few_shot_output_ollama}")
```

**Explanation of output (llama2)**: With several examples, the `llama2` model can better infer the structure for extracting information, leading to a more accurate and consistent output for the new address.

### Chain-of-Thought Prompting with Ollama (llama2)

**Concept**: This technique involves prompting the model to explicitly show its intermediate reasoning steps, or "chain of thought," before providing the final answer. This can significantly improve performance on complex reasoning tasks by allowing the model to break down the problem into smaller, more manageable steps.

**How it works with Ollama**: We'll instruct `llama2` to "think step by step" in the prompt, encouraging it to articulate its reasoning process before giving the final answer to a multi-step problem.

```python
cot_prompt_ollama = (
    "I have 3 apples. I buy 2 more apples. How many apples do I have now?\n"
    "Let's think step by step."
)

# Call Ollama model
response = ollama.chat(model='llama2', messages=[{
    'role': 'user',
    'content': cot_prompt_ollama,
}])
cot_output_ollama = response['message']['content']

print(f"--- Prompt ---\n{cot_prompt_ollama}\n")
print(f"--- Model Output (llama2) ---\n{cot_output_ollama}")
```

**Explanation of output (llama2)**: By guiding `llama2` to "think step by step," we encourage it to perform the calculation in stages, making the reasoning transparent and often leading to a more accurate answer for the arithmetic problem.
